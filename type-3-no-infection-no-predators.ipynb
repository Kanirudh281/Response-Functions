```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
def pom(H, t, r, K):
    return r * H * (1 - H/K)

r = 0.8    
K = 100.0  
t = np.linspace(0, 20, 1000) 
initial_conditions = [10, 50, 150] 

plt.figure(figsize=(10, 6))

for H0 in initial_conditions:
    sol = odeint(pom, H0, t, args=(r, K))
    plt.plot(t, sol, label=f'Initial Prey $H(0)={H0}$')

plt.axhline(y=K, color='r', linestyle='--', alpha=0.7, label=f'Carrying Capacity ($K={K}$)')
plt.title('Prey Population Dynamics (No Predators, No Infection)', fontsize=14)
plt.xlabel('Time ($t$)', fontsize=12)
plt.ylabel('Prey Population Density $H(t)$', fontsize=12)
plt.legend(loc='lower right')
plt.grid(True, linestyle=':', alpha=0.6)
plt.ylim(0, 160) 
plt.show()
```
