```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def model(z, t, r, K, a, b, c, d):
    x, y = z
    
    predation = (a * x**2) / (b**2 + x**2)
    dxdt = r * x * (1 - x / K) - predation * y
    dydt = c * predation * y - d * y
    return [dxdt, dydt]


r = 1.0    
K = 50      
a = 1.0     
b = 10    
c = 0.5  
d = 0.4   
x0 = 10     
y0 = 5     

t = np.linspace(0, 100, 1000)
solution = odeint(model, [x0, y0], t, args=(r, K, a, b, c, d))
x = solution[:, 0]
y = solution[:, 1]

plt.figure()
plt.plot(t, x, label="Prey (x)")
plt.plot(t, y, label="Predator (y)")
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Holling Type III Predator-Prey Model")
plt.legend()
plt.grid()
plt.show()
```
